## Closed Form Solution of Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
#create dataset
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

Y=Y.reshape((-1,1))
print(X.shape)
print(Y.shape)
```

```python
#normalize
X=(X-X.mean())/X.std()
```

```python
#visualize
plt.style.use('seaborn')
plt.figure()
plt.scatter(X,Y)
plt.title('Normalized Data')
plt.show()
```

```python
ones=np.ones((X.shape[0],1))
X_=np.hstack((X,ones))
print(X_.shape)
print(X_[:5,:])
```

```python
def pred(X,theta):
    return np.dot(X,theta)

def getThetaUsingClosedFormSolution(X,Y):
    Y=np.mat(Y)
    firstPart=np.dot(X.T,X)
    secondPart=np.dot(X.T,Y)
    
    theta=np.linalg.pinv(firstPart)*secondPart
    return theta

```

```python
theta=getThetaUsingClosedFormSolution(X_,Y)
print(theta)
```

```python
Y_=pred(X_,theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,Y_,color="red",label="prediction")
plt.legend()
plt.title('Normalized Data')
plt.show()
```
